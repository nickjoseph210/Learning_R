```r
# Example 1:
my_numeric = 0.2

print(class(my_numeric))
```

### FACTORS

**There is a FOURTH and more specialized Data Type: "factors."  They are seen as the categorical values we'd use in statistics.**

- Factors are text (or numbers) seen as a categorical value (male or female, true or false, etc)

```r
# Example 4: 
my_factor = as.factor("female") # note: you have to use 'as.factor'

print(class(my_factor))
```

### VECTORS

- A datatype that's a bunch of data types together

- You have to put a 'c' in front of the vector you're creating - it is short for 'concatenate'

```r
# Examples

vector_1 <- c(1, 2, 3, 4, 6)
print(class(vector_1))
```

```r
vector_2 = c("billy", "joe", "jack")
print(class(vector_2))
```

```r
vector_3 = c(T, F, FALSE, TRUE)
print(class(vector_3))
```

```r
vector_4 = c("billy", "joe", "jack", "22", "I love cowboy mouth")
print(class(vector_4))
```

**TO MAKE A VECTOR OF FACTORS:**

```r
vector_5 = as.factor(c(vector_1, vector_4)) # wrap the stuff you want to concatenate with 'as.factor()'
print(class(vector_5))
```

**You can put different datatypes together within the same vector, but there are a couple of rules:**

1.) If you put only 'TRUE' and 'FALSE' together, they will only be "true" and "false"; 

2.) If you put 'TRUE' and 'FALSE' with numeric values, 'TRUE' and 'FALSE' become numeric - 'true' = 1, 'false' = 0; &

3.) If you put numeric with boolean AND character values, EVERYTHING gets transformed to characters

```r
# Example
vector_6 = c(1, TRUE, "b")
print(class(vector_6))

vector_7 = c(1, 3, TRUE, FALSE)
print(class(vector_7))

vector_8 = c(3, 4, 5, 6)
print(class(vector_8))
```

**^^Seems like character changes everything.  A little life lesson there, for sure...**
